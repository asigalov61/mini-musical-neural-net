```python
import sys
import time

epoch = 0
batch = 0

for i in range(10):
    epoch +=1 
    batch +=100
    sys.stdout.write("\rEpoch: {0} Batch: {0}".format(epoch, batch))
    sys.stdout.flush()
    time.sleep(0.5)
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

x = [100]
y = [-100]

fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

while x[-1]>10 or x[-1]<-10:
    next_x = x[-1]*-0.9
    x.append(next_x)
    next_y = y[-1]*-0.9
    y.append(next_y)
    ax.clear()
    ax.plot(x)
    ax.plot(y)
    fig.canvas.draw()
    time.sleep(0.5)
```
